## 📖 PART 1: Foundations — Understanding Package Management

---

### 1.1 Why Do We Need Package Managers?

Python comes with built-in functions like `print()`, `input()`, `for`, and `if`. But what if you need:

```python
pandas      # Data manipulation
numpy       # Numerical computing
matplotlib  # Visualization
torch       # Deep learning
transformers # NLP models
openai      # LLM APIs
```

These libraries are **not included** with Python. Without package managers, you would have to:

```
Developer → Search Website → Download ZIP → Extract Files → Configure Paths → Pray It Works 😱
```

**The solution:**

```
Developer → pip install pandas → Done ✅
```

**Mental Model:** Package managers are the **App Store for Python**. PyPI (Python Package Index) at [pypi.org](https://pypi.org) is the repository containing 400,000+ packages.

---

### 1.2 What is pip?

**pip** (Pip Installs Packages) is Python's **official standard package installer**. It comes pre-installed with Python (since Python 3.4+) and has been the default for over 15 years.

**How pip works:**

```
pip install numpy
       ↓
  Contact PyPI (pypi.org)
       ↓
  Download Package
       ↓
  Install to site-packages/
       ↓
  Ready to Import ✅
```

**pip Strengths:**

| Strength | Description |
|----------|-------------|
| **Ubiquitous** | Installed with every Python distribution |
| **Simple** | Intuitive commands (`pip install`, `pip freeze`) |
| **Ecosystem** | Access to 400,000+ packages on PyPI |
| **Well-Documented** | 15+ years of community support |
| **Cross-Platform** | Works identically on Windows, macOS, Linux |

**pip Weaknesses:**

| Issue | Impact |
|-------|--------|
| **Slow** | Sequential dependency resolution causes long installs |
| **No Lock Files** | No built-in mechanism for reproducible builds |
| **Single-Purpose** | Only installs packages; needs separate tools for venv, pyenv, pip-tools, poetry |
| **Limited Caching** | Re-downloads packages for each virtual environment |
| **Global by Default** | Installs globally unless you manually manage venvs |

---

### 1.3 What is uv?

**uv** is an extremely fast Python package and project manager written in **Rust** by Astral (creators of the Ruff linter). Released in 2024, it's designed as a **drop-in replacement** for the entire fragmented Python toolchain:

```
pip        → Package installation
venv       → Virtual environments
pip-tools  → Dependency locking
pyenv      → Python version management
pipx       → Global tool installation
poetry     → Project management
```

**How uv works:**

```
uv add numpy
       ↓
  Parallel Resolution (Rust-powered)
       ↓
  Global Cache Check
       ↓
  Atomic Installation
       ↓
  Update pyproject.toml + uv.lock
       ↓
  Installed + Locked ✅
```

**uv Strengths:**

| Strength | Description |
|----------|-------------|
| **Blazing Fast** | 10-100x faster than pip (Rust, parallel downloads) |
| **Smart Resolution** | Modern parallel dependency resolver |
| **Lock Files** | Built-in `uv.lock` for reproducible builds |
| **Global Caching** | Single cache shared across all projects |
| **All-in-One** | Replaces pip, venv, pyenv, pip-tools, pipx, poetry |
| **Python Management** | Can install and manage Python versions directly |
| **Disk Efficient** | Uses hard links — 10 projects = only 1 copy on disk |
| **No Activation Needed** | `uv run` handles environment automatically |

**uv Weaknesses:**

| Issue | Impact |
|-------|--------|
| **Newer** | Less community adoption (2024+) |
| **Not Default** | Requires separate installation |
| **Learning Curve** | New commands to learn |

## 📖 PART 4: The Complete uv Workflow — Beginning to End

This is the most important section. Follow this exact workflow from installation to project completion.

---

### Step 0: Install uv

**Windows (PowerShell):**
```powershell
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

**macOS/Linux:**
```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
```

**Alternative (using pip):**
```bash
pip install uv
```

**Verify installation:**
```bash
uv --version
```

---

### Step 1: Create and Initialize a Project

```bash
# Create project directory:
mkdir llm-project
cd llm-project

# Initialize the project:
uv init
```

**What this creates:**
```
llm-project/
├── pyproject.toml      # Project configuration
├── main.py             # Default starter script
├── README.md           # Project description
└── .python-version     # Python version pin
```

---

### Step 2: Choose and Pin Python Version

```bash
# See available Python versions:
uv python list

# Install a specific Python version:
uv python install 3.12

# Pin this version for your project:
uv python pin 3.12
```

This creates/updates `.python-version` in your project root.

---

### Step 3: Add Dependencies

```bash
# Core data science packages:
uv add pandas numpy matplotlib seaborn

# AI/ML packages:
uv add scikit-learn

# LLM packages:
uv add openai google-genai sentence-transformers

# Development dependencies:
uv add --dev jupyter pytest black ruff
```

**What happens automatically:**
- `pyproject.toml` is updated with your dependencies
- `uv.lock` is generated with exact versions
- `.venv/` is created with all packages installed

---

### Step 4: Check Your Project State

```bash
# View the dependency tree:
uv tree

# View installed packages:
uv pip list

# Check for dependency conflicts:
uv pip check
```

---

### Step 5: Run Your Code

**The modern way (no activation needed):**
```bash
# Run a Python script:
uv run python main.py

# Run Jupyter Lab:
uv run jupyter lab

# Run Jupyter Notebook:
uv run jupyter notebook

# Run tests:
uv run pytest

# Format code:
uv run black .

# Lint code:
uv run ruff check .
```

**The traditional way (manual activation):**
```bash
# Windows:
.venv\Scripts\activate

# macOS/Linux:
source .venv/bin/activate

# Now run normally:
python main.py
```

---

### Step 6: Add More Dependencies Later

```bash
# Add new packages anytime:
uv add plotly dash
uv add --dev pre-commit

# No need to reactivate or reinstall — uv sync handles everything:
uv sync
```

---

### Step 7: Update Dependencies

```bash
# Update all packages to latest compatible versions:
uv lock --upgrade

# Update a specific package:
uv add pandas@latest

# Sync after updates:
uv sync
```

---

### Step 8: Save Your Work for GitHub

```bash
# Initialize git:
git init

# Create .gitignore:
echo ".venv/" >> .gitignore
echo "__pycache__/" >> .gitignore
echo "*.pyc" >> .gitignore

# Commit your work:
git add .
git commit -m "Initial project setup with uv"
git push
```

**Important:** Always commit these files:
- `pyproject.toml` — Your dependency declaration
- `uv.lock` — Exact versions for reproducibility
- `.python-version` — Python version specification

**Never commit:**
- `.venv/` — The virtual environment (can be recreated)

---

### Step 9: Deactivate the Environment (If Manually Activated)

```bash
# If you manually activated the venv:
deactivate

# Your terminal prompt returns to normal:
# user@machine:~/llm-project$
```

**With the modern `uv run` workflow, you don't need to deactivate — just close your terminal or VS Code when done.**

---

### Step 10: Reopen and Recreate the Project Later

```bash
# Clone the repository:
git clone <your-repo-url>
cd llm-project

# Install uv if needed:
curl -LsSf https://astral.sh/uv/install.sh | sh

# Recreate the exact environment:
uv sync

# Everything is now identical to your original setup!
uv run python main.py
```

---

### Step 11: Clean Up (Optional)

```bash
# Remove the virtual environment:
rm -rf .venv

# Clear uv global cache:
uv cache clean

# Check cache size:
uv cache dir
```

## 📖 PART 10: Interview Questions

**Q1: What is uv and how is it different from pip?**
A: uv is a modern Python package manager written in Rust. It's 10-100x faster than pip, has built-in lock files (`uv.lock`), manages virtual environments and Python versions, and uses a global cache for disk efficiency. It replaces pip, venv, pyenv, pip-tools, poetry, and pipx in a single tool.

**Q2: Why would you choose uv over pip?**
A: For speed (CI/CD pipelines), reproducible builds (lock files), team collaboration, disk efficiency (global cache with hard links), and unified toolchain.

**Q3: Is uv a drop-in replacement for pip?**
A: Yes. `uv pip install package` works identically to `pip install package`.

**Q4: What is a lock file and why is it important?**
A: A lock file (`uv.lock`) records exact versions of all dependencies and their dependencies. This ensures every developer and production environment uses identical packages, preventing "works on my machine" problems.

**Q5: Do I need to activate a virtual environment with uv?**
A: No. The modern uv workflow uses `uv run` which automatically handles the environment. This is the recommended approach by Astral. Manual activation (`source .venv/bin/activate`) is still supported but unnecessary for most workflows.
